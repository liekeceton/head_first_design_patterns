```python
import csv

csv_data = 'exam_results.csv'

class ExamResult:
    def __init__(self, exam_name, candidate_id, score, grade):
        self.exam_name = exam_name
        self.candidate_id = candidate_id
        self.score = score
        self.grade = grade

def get_course_summary(exam_name, exam_results):
    candidate_list = []
    score_list = []
    passed_exams = 0
    failed_exams = 0
    
    for result in exam_results:
        if result.exam_name == exam_name:
            candidate_list.append(result.candidate_id)
            score_list.append(result.score)
            
            if result.grade == 'Fail':
                failed_exams += 1
            else:
                passed_exams += 1
    
    results_summary = [
        exam_name, 
        len(set(candidate_list)), 
        passed_exams, failed_exams, 
        max(score_list), 
        min(score_list)
        ]
        
    return results_summary
    
#Exam results to list of ExamResults for practicality
exam_results = []
with open(csv_data, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        exam_results.append(ExamResult(row['Exam Name'], row['Candidate ID'], row['Score'], row['Grade']))

#Collect exam names
exam_names = []
for result in exam_results:
    exam_names.append(result.exam_name)
exam_names = list(set(exam_names))

#Create final report
import csv

with open('final_report.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    
    writer.writerow(
        ['Exam Name',
        'Number of Candidates',
        'Number of Passed Exams',
        'Number of Failed Exams',
        'Best Score',
        'Worst Score'])
        
    for exam_name in exam_names:
        course_summary = get_course_summary(exam_name, exam_results)
        writer.writerow(course_summary)
    
with open('final_report.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(row)
```
