```python
#Create a calculator
import random
import tkinter as tk
from tkinter import messagebox
```

```python
#I am stuck..
#LAB 3: THE CLICKER
nr_rows = 5
nr_columns = 5

#Do I need to create a button list?
button_nr = -1
button_list = []

#Create a main window
window = tk.Tk()
window.title('The clicker')

#Initialse timer variable
timer = tk.StringVar()
timer.set('0')

#Create handler for disabling a button when it is clicked
def clicked(idx):
    button = button_list[idx]
    button.config(state='disabled')
    
    #TODO: ONLY CHANGE TO DISABLED IF LOWEST NUMBER AVAILABLE!

def add_second():
    seconds_passed = int(timer.get()) + 1
    timer.set(str(seconds_passed))
    counter.after(1000, add_second)
    
#Create handler for starting the counter
def start_counter(ev):
    #Unbind after first event, for unbinding only indicate the event
    window.unbind_all('<Button-1>') 
    counter.after(1000, add_second)

#Create handles for stopping the counter
#Check if all buttons are disabled

#Sample to exclude the chance of double numbers
numbers = random.sample(range(1000), nr_rows*nr_columns) 

#Create the buttons
for row in range(nr_rows):
    for column in range(nr_columns):
        button_nr = button_nr + 1 #Index the buttons      
        number = numbers[button_nr] #Get the corresponding number for the button text
        
        #Create the button itself
        new_button = tk.Button( 
            text = str(number), 
            width = 10, 
            height = 3,
            command = lambda name = button_nr : clicked(name)) #A lambda function to the rescue..
        
        new_button.grid(row = row, column = column) #Locate the buttons
        button_list.append(new_button) #Append button to button list
        
#Create window for counter      
counter = tk.Label(textvariable=timer)
counter.grid(row=nr_rows+1, columnspan=nr_columns)

#Start the counter when anywhere on the window is clicked
window.bind_all('<Button-1>', start_counter) #This should work only once..

window.mainloop()
```

```python
#LAB 2: CATCH ME IF YOU CAN!
pad = 25 #Pixels padding around the button

#Create a main window
window = tk.Tk()
window.title('Catch me if you can!')
window.geometry("500x500")

#Create a function for the button to move if you move you mouse over it
def mouseover(ev): #Bind action when mouse is over
    random_x = random.randint(0 + pad, 500 - pad)
    random_y = random.randint(0 + pad, 500 - pad)
    button.place(x=random_x, y=random_y)
    
#Create a button to catch
button = tk.Button(window, text='Catch me!')
button.place(x=10, y=10) #Specify starting location

#Bind a command to the action of moving move over the button
button.bind("<Enter>", mouseover)

window.mainloop()
```

```python
#LAB 1: CREATE A CALCULATOR

#Create main window
window = tk.Tk()
window.title = 'Calculator'

#Create integer variabel + switch for switching between radiobuttons
switch = tk.IntVar()
switch.set(1)

#Create 
value_1 = tk.DoubleVar()
value_2 = tk.DoubleVar()

#Create command for evaluation
def clicked(): 
    #Get error messages when entries are not floats
    try:
        value_1_float = value_1.get()
    except:
        messagebox.showerror("ERROR", "First entry is not a number")
        entry_1.focus_set() #Set focus to window to be adapted
        return 
    
    try:
        value_2_float = value_2.get()
    except:
        messagebox.showerror("ERROR", "Second entry is not a number")
        entry_2.focus_set() #Set focus to window to be adapted
        return
    
    #Calculate evaluation
    if switch.get() == 0:
        calc = value_1.get() + value_2.get()
    elif switch.get() == 1:
        calc = value_1.get() - value_2.get()
    elif switch.get() == 2:
        calc = value_1.get() * value_2.get()
    elif switch.get() == 3:
        if value_2.get() == 0.0:
            #Do not allow zero division
            messagebox.showerror("ERROR", "Zero division is not allowed")
            return
        calc = value_1.get() / value_2.get()
        
    messagebox.showinfo("Evaluation", str(calc))

#Create widgets
entry_1 = tk.Entry(window, textvariable=value_1)
entry_2 = tk.Entry(window, textvariable=value_2)
rb_1 = tk.Radiobutton(window, text='+', variable=switch, value=0)
rb_2 = tk.Radiobutton(window, text='-', variable=switch, value=1)
rb_3 = tk.Radiobutton(window, text='*', variable=switch, value=2)
rb_4 = tk.Radiobutton(window, text='/', variable=switch, value=3)
button = tk.Button(window, text='Evaluate', command=clicked)

#Locate widgets
entry_1.grid(column=0, row=2)
entry_2.grid(column=2, row=2)
rb_1.grid(column=1, row=0)
rb_2.grid(column=1, row=1)
rb_3.grid(column=1, row=2)
rb_4.grid(column=1, row=3)
button.grid(column=1, row=4)

#Start main loop (obligatory)
window.mainloop()
```
